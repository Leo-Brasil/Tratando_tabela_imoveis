### Produto para imovel residencial

```python
# Como o cliente quer lançar um produto apenas para imovel residencial, iremos criar um df apenas com os imovei residenciais
# Carregamos a nossa base de dados para fica mais visivel
aluguel
```

```python
# Vamos ver quais são os tipos de imoveis que temos
aluguel['Tipo']
```

```python
# Fazemos umas lista dela e tiramos as duplicatas
list(aluguel['Tipo'].drop_duplicates())
```

```python
# Copiamos a lista acima e selecionamos manualmente apenas as que são residenciais, atribuimos uma variavel a ela
residencial = ['Quitinete', 'Casa', 'Apartamento', 'Casa de Condomínio', 'Casa de Vila']
```

```python
# Com o metodo .isin() usamos para listar True ou False na coluan tipo no nosso Data Frame 'Aluguel', atribuimos a variavel
selecao = aluguel['Tipo'].isin(residencial)
```

```python
# Retorna uma lista booleana
selecao
```

```python
# Com a variavel atribuida, jogamos ela entre colchetes, para que ela criasse um dataframe apenas com os imoveis residenciais
aluguel_residencial = aluguel[selecao]
aluguel_residencial
```

```python
# Verificamos os imoveis, para ver se a seleçao e processo estão corretos
list(aluguel_residencial['Tipo'].drop_duplicates())
```

```python
# Visto que esta correto a formula acima, vamos arrumar o index, ajustando com o tamanho do nosso novo DF
aluguel_residencial.shape[0]
```

```python
# Visualizando a cabeça, apenas para verificar como o index está
aluguel_residencial.head(10)
```

```python
# Criamos o index com o shape
index = aluguel_residencial.shape[0]
```

```python
# Aplicamos a variamos index junto com o range, para que nosso index fique de forma crescenete em ordem
aluguel_residencial.index = range(index)
aluguel_residencial
```

```python
# Exportamos a base de dados que nosso cliente quer, para futuras analises
aluguel_residencial.to_csv('dados/aluguel_residencial.csv', sep = ';', index = False)
```

```python
# Vemos se a base de dados subiu de forma correta
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
dados
```

### Seleção e Frequencia

```python
# Selecione somente os imóveis classificados com tipo 'Apartamento'.
Apartamentos = dados['Tipo'] == 'Apartamento'
n1 = dados[Apartamentos].shape[0]
```

```python
# Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.
Casas = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
n2 = dados [Casas].shape[0]
```

```python
# Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.
Area = (dados['Area'] >= 60) & (dados['Area'] <= 100)
n3 = dados[Area].shape[0]
```

```python
# Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.
selecao_de_imoveis = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000)
n4 = dados[selecao_de_imoveis].shape[0]
```

```python
print(f"Número de imóveis classificados com tipo 'Apartamento' -> {n1}.")
print(f"Número de imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila' -> {n2}.")
print(f"Número de imóveis com área entre 60 e 100 metros quadrados, incluindo os limites -> {n3}.")
print(f"Número de imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00 -> {n4}.")
```

### Retirando dados nulos 

```python
aluguel_residencial
```

```python
''' Vemos que temos 22580 registros, e apenas 22571 de valor, 20765 de condominio e  apenas 15795 de IPTU'''
aluguel_residencial.info()
```

```python
# Faremos uma matriz booleana 
aluguel_residencial['Valor'].isnull()
```

```python
# Colocaremos em um dataframe os valores que apareceram True na matriz booleana acima.
aluguel_residencial[aluguel_residencial['Valor'].isnull()]
```

```python
# Como o valor do Aluguel é a nossa principal função do Datafram, não faz sentido eu ter imoveis sem esse valor
# Por isso iremos fazer um dropna()
# Repare que aqui o dropna, precisa ser passado junto com o inplace
'''Utilizamos dados que é o mesmo dataframe, mas como colocaremos o inplace irá substituir o aluguel_residencial, 
assim conseguiremos fazer a subtração para ver quantos registros foram excluidos do nosso Dataframe'''
aluguel_residencial.dropna(subset =['Valor'], inplace = True)
```

```python
# Agora vemos que nosso DataFrame possui 22571 registros
aluguel_residencial
```

```python
aluguel_residencial.info()
```

```python
# Agora resolver o problema do Condominio
# Vamos ver quantos valores nulos temos na coluna 'Condominio'
'''Não é porque o valor é nulo, que o valor esta errado, por exemplo casas não possuem tarifas de condominio, já é diferente
de apartamentos que temos certeza que possui valor de condominio'''
aluguel_residencial[aluguel_residencial['Condominio'].isnull()].shape[0]
```

```python
# Fazendo seleção de Tipo de imovel como apartamento e que possuem condominio nulo, pois sabemos que apto. possui condominio, 
# então iremos retirar do nosso dataframe
selecao = (aluguel_residencial['Tipo'] == 'Apartamento') & (aluguel_residencial['Condominio'].isnull())
```

```python
# Vemos a quantidade de linha no nosso aluguel_residencial antes de eliminar nossa selecao
aluguel_residencial.shape[0]
```

```python
# aplicamos a seleção mas de forma contraria, deixando o restante e eliminando apenas a seleção
aluguel_residencial = aluguel_residencial[~selecao]
```

```python
# Vemos que retiramos a seleção do nosso data frame
aluguel_residencial.shape[0]
```

```python
# De 1813 fomos para 1068 resultados nulos em condominio, agora como eliminamos os apto. so restam casas com campo nulo.
# Vamos tratar essas casas que realmente não possuem condominio atribuindo valor 0 a elas.
aluguel_residencial[aluguel_residencial['Condominio'].isnull()].shape[0]
```

```python
# Agora que retiramos os apto. que possuiam condominio nulos, iremos tratar as casas que possuem valores nulos, substituindo por 0.
aluguel_residencial.fillna({'Condominio': 0, 'IPTU': 0}, inplace = True)
```

```python
aluguel_residencial
```

```python
# Agora que substituimos os valores nulos do IPTU e condominio, vamos verificar se esta certo com o info()
aluguel_residencial.info()
```
